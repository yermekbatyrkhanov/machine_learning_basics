```python
import matplotlib.pyplot as plt
import mglearn
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Example data
X, y = make_blobs(random_state=1)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# First plot: KMeans with 2 clusters
kmeans2 = KMeans(n_clusters=2, random_state=0)
kmeans2.fit(X)
assignments2 = kmeans2.labels_
mglearn.discrete_scatter(X[:, 0], X[:, 1], assignments2, ax=axes[0])
axes[0].set_title("KMeans with 2 clusters")

# Second plot: KMeans with 5 clusters
kmeans5 = KMeans(n_clusters=5, random_state=0)
kmeans5.fit(X)
assignments5 = kmeans5.labels_
mglearn.discrete_scatter(X[:, 0], X[:, 1], assignments5, ax=axes[1])
axes[1].set_title("KMeans with 5 clusters")

# Show both plots together
plt.show()
```

```python
X_varied, y_varied = make_blobs(n_samples = 200, cluster_std=[1.0, 2.5,0.5],
                               random_state= 170)
```

```python
y_pred = KMeans(n_clusters=3, random_state=0).fit_predict(X_varied)
```

```python
mglearn.discrete_scatter(X_varied[:,0], X_varied[:,1], y_pred)
plt.legend(["cluster 0", "cluster 1", "cluster 2"], loc ='best')
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
import numpy as np 
rng = np.random.RandomState(74)
```

```python
print(rng)
```

```python
#transform the data to be stretched
transformation = rng.normal(size=(2,2))
```

```python
X=np.dot(X,transformation)
```

```python
kmeans=KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
y_pred=kmeans.predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=y_pred, cmap=mglearn.cm3)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
           marker='^', c=[0,1,2], s=100, linewidth=2, cmap=mglearn.cm3)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
from sklearn.datasets import make_moons
X,y = make_moons(n_samples = 200, noise = 0.05, random_state =0)

# cluster the data into two clusters 
kmeans = KMeans(n_clusters = 2)
kmeans.fit(X)
y_pred = kmeans.predict(X)
```

```python
# plot the cluster assignments and cluster centers 
plt.scatter(X[:,0], X[:,1], c=y_pred, cmap=mglearn.cm2, s=60)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
           marker ='^', c=[mglearn.cm2(0), mglearn.cm2(1)], s=100, linewidth=2)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```
